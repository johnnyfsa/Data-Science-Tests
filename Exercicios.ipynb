```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Conjunto de dados TechTaste
df_techtaste = pd.DataFrame({'avaliacoes': [38, 44, 33, 42, 47, 33, 36, 39, 42, 36, 39, 34, 42, 42, 36, 43, 31, 35, 36, 41, 42, 30, 25, 38, 47, 36, 32, 45, 44, 45, 37, 48, 37, 36, 44, 49, 31, 45, 45, 40, 36, 50, 38, 34, 36, 42, 46, 49, 36, 34, 38, 31, 53, 40, 57, 40, 36, 42, 26, 50, 32, 43, 35, 37, 42, 30, 36, 43, 40, 43, 44, 52, 37, 51, 35, 47, 40, 50, 37, 49]})
```

```python
desvio_padrao = df_techtaste.std()
n = len(df_techtaste)
EP = desvio_padrao/np.sqrt(n)
media = df_techtaste.mean()
#gerar valores minimos e maximos para x
x_min=min(df_techtaste['avaliacoes'])
x_max=max(df_techtaste['avaliacoes'])
x = np.linspace(x_min, x_max,n)
distribuicao_normal = stats.norm.pdf( x, media, desvio_padrao)
```

```python
fig,ax = plt.subplots()
ax.hist(df_techtaste, linewidth=0.5, edgecolor='white')
ax.plot(x, distribuicao_normal,'k', label='Distribuicao Normal', linewidth=1.0, color='red')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Plotar o histograma no primeiro eixo
ax1.hist(df_techtaste, bins=25, density=True, alpha=0.5, label='Histograma')

# Plotar a distribuição normal no segundo eixo
ax2.plot(x, distribuicao_normal, 'r-', label='Distribuição Normal')

# Personalizar os gráficos
ax1.set_title('Histograma e Distribuição Normal')
ax1.set_ylabel('Densidade')
ax2.set_xlabel('Valores')
ax2.set_ylabel('Densidade')

# Adicionar legendas
ax1.legend()
ax2.legend()

plt.tight_layout()
plt.show()
```
